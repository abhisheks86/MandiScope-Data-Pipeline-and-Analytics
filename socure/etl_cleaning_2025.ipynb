```python
import pandas as pd 
df = pd.read_csv("raw_2025.csv")
display(df.head())
print(df.shape)
print(df.dtypes)

null_counts = df.isna().sum()
null_percentages = (null_counts/len(df))*100
missing_data = pd.DataFrame({'Missing_Rows':null_counts,'Missing_%':null_percentages})

print(missing_data[missing_data['Missing_Rows'] > 0].sort_values(by='Missing_Rows', ascending=False))

numeric_cols = ['Min_Price','Max_Price','Modal_Price']
print(df[numeric_cols].describe().T[['min','max','mean','50%']])
math = df[df['Min_Price']>df['Max_Price']]
print(f"\nAlert: Found {len(math):,} rows where Min_Price > Max_Price!")
zero_prices = df[df['Modal_Price']<=0]
print(f"Alert:found{len(zero_prices):,}rows with a Model_Price of 0 or negative!")


for col in ['STATE','Commodity']:unique_count = df[col].nunique()
print(f"\n column [{col}] has {unique_count}distinct spellings.")
print(df[col].value_counts().head(10))

#Clean
text_cols = ['STATE', 'District Name', 'Market Name', 'Commodity', 'Variety', 'Grade']
for col in text_cols:
    df[col] = df[col].fillna("UNKNOWN")
    df[col] = df[col].astype(str).str.strip().str.upper()
df = df.dropna(subset=['Modal_Price', 'Min_Price', 'Max_Price'])
df = df[(df['Min_Price'] > 0) & (df['Max_Price'] > 0) & (df['Modal_Price'] > 0)]
df = df[df['Min_Price'] < 100000]
df = df[df['Max_Price'] >= df['Min_Price']]

#verify 
print(f"Cleaned Row Count: {len(df):,}")
rows_removed = 135358 - len(df) 
print(f"Total Garbage Rows Purged: {rows_removed:,}")
print(df[['Min_Price', 'Max_Price', 'Modal_Price']].describe().T[['min', 'max']])

null_check = df.isna().sum()
remaining = null_check[null_check > 0]
if len(remaining) == 0:
    print("Zero nulls remaining — data is clean")
else:
    print(remaining)
bad_math = df[df['Min_Price'] > df['Max_Price']]
print(f"Min > Max remaining: {len(bad_math)}")

#Save 
df.to_csv("Clean_Mandi_2025.csv", index=False)
print("Saved file")
```
