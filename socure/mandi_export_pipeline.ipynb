```python
import pandas as pd
from sqlalchemy import create_engine

print("PART 1: Fixing Mandi_All_Years_2023_2026.csv")

df = pd.read_csv("Mandi_All_Years_2023_2026.csv")

print("Columns:", df.columns.tolist())
print("\nBefore fix — null check:")
print(df.isna().sum())

df['Price Date'] = pd.to_datetime(df['Price Date'], dayfirst=True, errors='coerce')
df['Price Date'] = df['Price Date'].fillna(pd.Timestamp('2026-01-01'))

print("\nAfter fix — null check:")
print(df.isna().sum())
print(f"\nTotal rows: {len(df):,}")

df.to_csv("Mandi_All_Years_2023_2026.csv", index=False)
print("Saved!")


print("\nPART 2: Exporting from MySQL")

engine = create_engine("mysql+pymysql://root:abhiguru46@localhost:3306/mandi")

df_fact = pd.read_sql("SELECT * FROM fact_mandi_prices", engine)
print(f"Rows loaded: {len(df_fact):,}")
df_fact.to_csv("fact_mandi_prices.csv", index=False)
print("fact_mandi_prices.csv saved!")

tables = ['dim_location', 'dim_date', 'dim_commodity']
for table in tables:
    try:
        df_dim = pd.read_sql(f"SELECT * FROM {table}", engine)
        df_dim.to_csv(f"{table}.csv", index=False)
        print(f"{table}.csv saved — {len(df_dim):,} rows")
    except Exception as e:
        print(f"{table} not found: {e}")

print("\nAll done!")
```
