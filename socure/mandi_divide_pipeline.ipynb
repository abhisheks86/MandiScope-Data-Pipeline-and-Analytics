```python
import pandas as pd

print("Initiating Pure Raw Data Partitioning (Bronze Layer)...\n")
file_name = "Agriculture_price_dataset_unclean_full_more_messy.csv"
df_raw = pd.read_csv(file_name)
date_col = 'Price Date'
df_raw['Temp_Year'] = pd.to_datetime(df_raw[date_col], errors='coerce').dt.year
unique_years = df_raw['Temp_Year'].dropna().unique()

print(f"Total raw rows loaded: {len(df_raw):,}")
print(f"Years found: {sorted([int(y) for y in unique_years])}")
print(f"Rows with no date: {df_raw['Temp_Year'].isna().sum():,}\n")
# ----------------------------------------

print("--- Outputting Raw Year Files ---")
for year in sorted(unique_years):
    df_year = df_raw[df_raw['Temp_Year'] == year].copy()
    df_year = df_year.drop(columns=['Temp_Year'])
    out_name = f"Raw_Mandi_Unclean_{int(year)}.csv"
    df_year.to_csv(out_name, index=False)
    
    print(f"Extracted: {out_name} -> {len(df_year):,} rows.")

df_no_date = df_raw[df_raw['Temp_Year'].isna()].copy()
if not df_no_date.empty:
    df_no_date = df_no_date.drop(columns=['Temp_Year'])
    df_no_date.to_csv("Raw_Mandi_Unclean_NO_DATE.csv", index=False)
    print(f"Extracted: Raw_Mandi_Unclean_NO_DATE.csv -> {len(df_no_date):,} rows with broken dates.")

print("\nRaw Partitioning Complete.")
```
